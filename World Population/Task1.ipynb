

# World Population Analysis using Python


```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
# Load dataset

df = pd.read_csv("/content/world_population.csv")

# Display first 5 rows

df.head()
```

```python
# Dataset shape

print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
# Dataset information

df.info()
```

```python
# Column names

df.columns
```

```python
# Missing values

df.isnull().sum()
```

```python
# Duplicate values

print(
    "Duplicate Rows :",
    df.duplicated().sum()
)
```

```python
# Summary statistics

df.describe()
```

```python
# Top 10 populated countries

top10 = df.sort_values(
    by="2022 Population",
    ascending=False
).head(10)

top10[
    [
        "Country/Territory",
        "2022 Population"
    ]
]
```

```python
# Top 10 countries by population

plt.figure(figsize=(12,5))

colors = [
    "royalblue",
    "tomato",
    "seagreen",
    "orange",
    "purple",
    "gold",
    "teal",
    "brown",
    "pink",
    "gray"
]

plt.bar(
    top10["Country/Territory"],
    top10["2022 Population"],
    color=colors,
    edgecolor="black"
)

plt.title("Top 10 Countries by Population")

plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.grid(axis="y", alpha=0.4)

plt.show()
```

```python
# Bottom 10 populated countries

bottom10 = df.sort_values(
    by="2022 Population"
).head(10)
```

```python
# Bottom 10 countries by population

plt.figure(figsize=(10,5))

colors = [
    "royalblue",
    "tomato",
    "mediumseagreen",
    "gold",
    "purple",
    "teal",
    "orange",
    "deeppink",
    "slateblue",
    "brown"
]

plt.barh(
    bottom10["Country/Territory"],
    bottom10["2022 Population"],
    color=colors,
    edgecolor="black"
)

plt.title("Bottom 10 Countries by Population")

plt.xlabel("Population")
plt.ylabel("Country")

plt.grid(axis="x", alpha=0.4)

plt.show()
```

```python
# Population Distribution Histogram

plt.figure(figsize=(10,5))

plt.hist(
    df["2022 Population"],
    bins=20,
    color="darkorange",
    edgecolor="black"
)

plt.title("Population Distribution")

plt.xlabel("Population")
plt.ylabel("Frequency")

plt.grid(alpha=0.4)

plt.show()
```

```python
# Continent Wise population

continent_pop = df.groupby(
    "Continent"
)["2022 Population"].sum()
```

```python
# Continent wise population

plt.figure(figsize=(10,5))

plt.bar(
    continent_pop.index,
    continent_pop.values,
    color=[
        "royalblue",
        "tomato",
        "mediumseagreen",
        "gold",
        "purple",
        "teal"
    ]
)

plt.title(
    "Continent Wise Population"
)

plt.xlabel("Continent")
plt.ylabel("Population")

plt.show()
```

```python
# Top 10 Densely Populated Countries

top_density = df.sort_values(
    by="Density (per km²)",
    ascending=False
).head(10)
```

```python
# Top 10 densely populated countries

plt.figure(figsize=(12,5))

plt.bar(
    top_density["Country/Territory"],
    top_density["Density (per km²)"],
    color=[
        "royalblue",
        "dodgerblue",
        "deepskyblue",
        "mediumseagreen",
        "limegreen",
        "gold",
        "orange",
        "darkorange",
        "tomato",
        "crimson"
    ],
    edgecolor="black"
)

plt.title("Top 10 Densely Populated Countries")

plt.xlabel("Country")
plt.ylabel("Density (per km²)")

plt.xticks(rotation=45)

plt.grid(axis="y", alpha=0.4)

plt.show()
```

```python
# Population vs Density

# Population vs Density

plt.figure(figsize=(10,5))

plt.scatter(
    df["Density (per km²)"],
    df["2022 Population"],
    c=df["Growth Rate"],
    cmap="viridis",
    s=80
)

plt.colorbar(label="Growth Rate")

plt.title("Population vs Density")

plt.xlabel("Density")
plt.ylabel("Population")

plt.grid(alpha=0.4)

plt.show()
```

```python
# Top Growth Rate Countries

top_growth = df.sort_values(
    by="Growth Rate",
    ascending=False
).head(10)
```

```python
# Top 10 countries by growth rate

plt.figure(figsize=(10,5))

plt.barh(
    top_growth["Country/Territory"],
    top_growth["Growth Rate"],
    color=[
        "royalblue",
        "dodgerblue",
        "deepskyblue",
        "mediumseagreen",
        "limegreen",
        "gold",
        "orange",
        "darkorange",
        "tomato",
        "crimson"
    ],
    edgecolor="black"
)

plt.title("Top 10 Countries by Growth Rate")

plt.grid(axis="x", alpha=0.4)

plt.show()
```

```python
# Population Trend Analysis

# Top 5 countries

top5 = df.sort_values(
    by="2022 Population",
    ascending=False
).head(5)
```

```python
years = [
    "1970 Population",
    "1980 Population",
    "1990 Population",
    "2000 Population",
    "2010 Population",
    "2015 Population",
    "2020 Population",
    "2022 Population"
]
```

```python
# Population growth trend

plt.figure(figsize=(12,5))

colors = [
    "royalblue",
    "tomato",
    "mediumseagreen",
    "darkorange",
    "purple"
]

for i in range(len(top5)):

    plt.plot(
        years,
        top5.iloc[i][years],
        marker="o",
        linewidth=2,
        color=colors[i],
        label=top5.iloc[i]["Country/Territory"]
    )

plt.title("Population Growth Trend")

plt.xlabel("Year")
plt.ylabel("Population")

plt.legend()

plt.grid(alpha=0.4)

plt.xticks(rotation=45)

plt.show()
```

```python
# Population Outlier Detection

# Violin plot

plt.figure(figsize=(8,5))

plt.violinplot(
    df["2022 Population"]
)

plt.title("Population Distribution")

plt.grid(alpha=0.3)

plt.show()
```

```python
# Correlation Heatmap
# Numerical columns

numeric_cols = [
    "2022 Population",
    "2020 Population",
    "2015 Population",
    "2010 Population",
    "Area (km²)",
    "Density (per km²)",
    "Growth Rate"
]
```

```python
# Correlation matrix

df[numeric_cols].corr()
```

```python
# Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title(
    "Correlation Heatmap"
)

plt.show()
```
